### 常用範例

```python
import matplotlib.pyplot as plt

month1=[1,2,3,4,5,6,10,12]
sale1=[20000,40000,60000,80000,100000,120000,140000,160000]

plt.plot(month1, sale1, color='red', linewidth=1.0, linestyle='--', marker='^', label='Product A')

month2=[1,3,4,5,6,7,11,12]
sale2=[10000,20000,30000,15000,120000,80000,60000,210000]

plt.plot(month2, sale2, color='blue', linewidth=1.0, linestyle=':', marker='o', label='Product B')

plt.xlim((0, 14))
plt.ylim((0, 250000))

plt.title('Profit')
plt.xlabel('Month')
plt.ylabel('Sale')
plt.legend()

plt.savefig("./example.png")

plt.show()
```

### Line Style

```python
import matplotlib.pylab as plt
linestyle=['-','--','-.',':']
descriptions=['solid line', 'dashed line', 'dash-dot line', 'dotted line']
x=[]
y=[]
for i in range(5):
    for j in range(5):
        x.append(i)
        y.append(j)
plt.figure()
# color='red', linewidth=1.0, linestyle='--', marker='^', label='Product A')
for i,j,l,d in zip(x,y,linestyle,descriptions):
    plt.plot([i, i+0.5], [j, j], linewidth=2.0, linestyle=l)
    plt.text(i-0.05, j+0.25, s=l+' '+d, fontdict= {'family':'monospace', 'weight':'bold', 'size':16})
plt.axis([-0.1, 1.8, -0.1, 4.5])
plt.tight_layout()
plt.axis('off')
plt.title('Line Style')
plt.show() 
```

### Mark Style

```python
import matplotlib.pylab as plt
markers=['.',',','o','v','^','<','>','1','2','3','4','8','s','p','P','*','h','H','+','x','X','D','d','|','_']
descriptions=['point', 'pixel', 'circle', 'triangle_down', 'triangle_up','triangle_left', 'triangle_right', 'tri_down', 'tri_up', 'tri_left','tri_right', 'octagon', 'square', 'pentagon', 'plus (filled)','star', 'hexagon1', 'hexagon2', 'plus', 'x', 'x (filled)','diamond', 'thin_diamond', 'vline', 'hline']
x=[]
y=[]
for i in range(5):
    for j in range(5):
        x.append(i)
        y.append(j)
plt.figure(figsize=(15, 5))
for i,j,m,d in zip(x,y,markers,descriptions):
    plt.scatter(i, j, marker=m, s=200)
    plt.text(i-0.05, j+0.15, s=m+':'+d, fontdict= {'family':'monospace', 'weight':'bold', 'size':16})
plt.axis([-0.1, 4.8, -0.1, 4.5])
plt.tight_layout()
plt.axis('off')
plt.show() 
```
